Площади домов расположились, в основном, в интервале от 250000 до 700000. При этом, так же как и со стоимостью домов, присутствует много выбросов.

```python
plt.figure(figsize=(10,6))
plt.hist(df['yr_built'], bins=30)
plt.locator_params (axis='x', nbins= 30)
plt.xlabel('год постройки')
plt.ylabel('количество домов')
plt.title('плотность распределения годов постройки');
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['yr_built'])
plt.locator_params (axis='y', nbins= 30)
plt.ylabel('год постройки')
plt.title('плотность распределения годов постройки')
plt.grid();
```

Года постройки домов расположились, в основном, в интервале от 1950 до 2000. Выбросов нет.

```python
df['waterfront'] = list(map(lambda item: 'yes' if item == 1 else 'no', df['waterfront']))
```

```python
df['waterfront'].value_counts()
```

```python
plt.figure(figsize=(4,6))
plt.bar(df['waterfront'].unique(), df['waterfront'].value_counts())
plt.xlabel('вид на набережную')
plt.ylabel('количество домов')
plt.title('зависимость  средней цены дома от вида из окна');
```

Вид на набережную имеют лишь 163 дома.

```python
plt.figure(figsize=(10,6))
# plt.bar(df['floors'].unique(), df['floors'].value_counts())
sns.barplot(x=df['floors'].unique(), y=df['floors'].value_counts())
plt.locator_params (axis='x', nbins= 10)
plt.xlabel('количество этажей')
plt.ylabel('количество домов')
plt.title('Этажность домов');
```

Основная часть домов имеет один или два этажа.

```python
plt.figure(figsize=(10,6))
# plt.plot(df.groupby('condition').agg({'price': ['mean']}))
sns.barplot(x=df['condition'].unique(), y=df['condition'].value_counts())
plt.xlabel('состояние дома')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от его состояния')
plt.grid();
```

Более половины домов имеет среднее техническое состояние.

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('bedrooms').agg({'price': ['mean']}))
plt.xlabel('количество спален')
plt.ylabel('средняя цена')
plt.title('зависимость срелней цены дома от количества спален')
plt.grid();
```

Количество спален оказывает влияние на среднюю цену дома, однако зависимость является не линейной. Этот параметр в меньшей степени влияет на цену дома.

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('bathrooms').agg({'price': ['mean']}))
plt.xlabel('количество ванных комнат')
plt.ylabel('средняя цена')
plt.title('зависимость средней цены дома от количества ванных комнат')
plt.grid();
```

Количество ванных комнат оказывает влияние на среднюю цену дома, эта зависимость близка к линейной. Влияние этого параметра влияет на стоимость дома, но скорее всего из-зи кореляции с другими параметрами, влияющими на цену дома в значительной степени.

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('floors').agg({'price': ['mean']}))
plt.xlabel('количество этажей')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от количества этажей')
plt.grid();
```

Зависимость цены дома от его этажности не наблюдается

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('condition').agg({'price': ['mean']}))
plt.xlabel('состояние дома')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от его состояния')
plt.grid();
```

Прослеживается почти линейная зависимость цены дома от его состояния.

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('view').agg({'price': ['mean']}))
plt.xlabel('вид из окна')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от вида из окна')
plt.grid();
```

Вид из окна почти линейно и  в значительной степени влияет на цену дома.

```python
plt.figure(figsize=(4,6))
sns.barplot(x=df['waterfront'].unique(), y=df.groupby('waterfront').agg({'price': ['mean']}))
plt.xlabel('вид на набережную')
plt.ylabel('количество домов')
plt.title('зависимость  средней цены дома от вида из окна');
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('grade').agg({'price': ['mean']}))
plt.xlabel('категория')
plt.ylabel('средняя цена')
plt.title('зависимость средней цены дома от категории')
plt.grid();
```
