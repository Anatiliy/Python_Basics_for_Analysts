```python
# считываем данные.
df = pd.read_csv('../../data/kc_house_data.csv')
# df['price'] = df['price'] / 1000
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('bedrooms').agg({'price': ['mean']}))
plt.xlabel('количество спален')
plt.ylabel('средняя цена')
plt.title('зависимость срелней цены дома от количества спален')
plt.grid();
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('bathrooms').agg({'price': ['mean']}))
plt.xlabel('количество ванных комнат')
plt.ylabel('средняя цена')
plt.title('зависимость средней цены дома от количества ванных комнат')
plt.grid();
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('floors').agg({'price': ['mean']}))
plt.xlabel('количество этажей')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от количества этажей')
plt.grid();
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('condition').agg({'price': ['mean']}))
plt.xlabel('количество этажей')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от количества этажей')
plt.grid();
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('view').agg({'price': ['mean']}))
plt.xlabel('количество этажей')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от количества этажей')
plt.grid();
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.groupby('waterfront').agg({'price': ['mean']}))
plt.xlabel('количество этажей')
plt.ylabel('средняя цена')
plt.title('зависимость  средней цены дома от количества этажей')
plt.grid();
```

```python
plt.figure(figsize=(20,6))
plt.plot(df['sqft_lot'].sort_values(), df['price'])
plt.xlabel('общая площадь')
plt.ylabel('цена')
plt.title('зависимость цены дома от его площади')
plt.grid();
```

```python
plt.figure(figsize=(20,6))
plt.hist(df['price'], bins=200)
plt.xlabel('цена')
plt.ylabel('частота распределения')
plt.title('распределение цены');
```
